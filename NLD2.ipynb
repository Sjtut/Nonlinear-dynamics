```python
import numpy as np
import matplotlib.pyplot as plt

NY=50 #delta_y=0.02
n=1000#number of iteration
for k in [0,0.2,0.7,1.1,1.5,0.9]:
    X=[]
    Y=[]
    for i in range (NY):
        x=0.5
        y=i/NY
        X.append(x)
        Y.append(y)
        for j in range(n-1):
            y=((y-k/2/np.pi*np.sin(2*np.pi*x))%1)# calculate Y_i+1
            x=((x+y)%1)# calculate X_i+1
            X.append(x)
            Y.append(y)
    plt.plot(X,Y,'r.',markersize=0.05)
    plt.title('k = %.1f' %k)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```
