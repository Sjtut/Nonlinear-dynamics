```python
import numpy as np
import matplotlib.pyplot as plt
NX=10 #delta_x=0.1
NY=40 #delta_y=0.1
n=500#number of iteration
a=0.68
for b in [0,0.7]:
    X=[]
    Y=[]
    for k in range (NY):
        for i in range (NX):
            #initialization
            x=i/NX
            y=k/NY*4-2
            X.append(x)
            Y.append(y)
            for j in range(n-1):
                y=(y-b*np.sin(2*np.pi*x))# calculate Y_i+1 
                x=(x+a*(1-y**2))%1# calculate X_i+1,x ranges(0,1)
                X.append(x)
                Y.append(y)
    plt.plot(X,Y,'r.',markersize=0.05)
    plt.title(''.join(['b=',str(b),',a=0.68']))
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

```python
#zoom in the center 
plt.plot(X,Y,'r.',markersize=0.05)
plt.title(''.join(['b=',str(b),',a=0.68']))
plt.xlabel('X')
plt.ylabel('Y')   
plt.axis([0.0,1.0,-2.0,2.0])
plt.show()
```
